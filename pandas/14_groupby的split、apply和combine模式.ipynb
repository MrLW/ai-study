### 归一化


```python
import pandas as pd
df_rattings=pd.read_csv('./data/ml-25m/ratings.csv')
```

```python
df_rattings["rating"].min()
```

```python
type(df_rattings.groupby(["userId"]))
```

```python
def ratting_norm(df):
    max_value=df["rating"].max()
    min_value=df["rating"].min()
    df["norm_rating"] = df["rating"].apply(lambda x: 0 if max_value == min_value else (x - min_value) / (max_value - min_value))
    return df
df_rattings.groupby(["userId"]).apply(ratting_norm)
```

```python
def getTopNData(df, topn):
    return df.sort_values(by="rating")[-topn:]
    
# 注意这里apply参数的应用
df_rattings.groupby(["userId"]).apply(getTopNData, topn=5).head(10)
```

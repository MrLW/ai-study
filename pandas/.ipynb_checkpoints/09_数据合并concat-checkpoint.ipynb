#### 场景使用
批量操作dataframe，给dataframe增加行、列

concat
append

```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    "A":["A1","A2","A3"],
    "B":["B1","B2","B3"],
    "C":["C1","C2","C3"],
})
df2=pd.DataFrame({
    "A":["A4","A5","A6"],
    "B":["B4","B5","B6"],
    "D":["D4","D5","D6"],
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
# ignore_index：是否重建索引
pd.concat([df1,df2], ignore_index=True)
```

```python
# 默认join 为outer
pd.concat([df1,df2], ignore_index=True, join="inner")
```

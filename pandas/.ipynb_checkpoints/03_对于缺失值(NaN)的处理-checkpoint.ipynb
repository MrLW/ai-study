```python
import pandas as pd
```

```python
studf=pd.read_excel('./data/student_dirty.xlsx', skiprows=2)
```

```python
# 打印出所有是否是null
studf.isnull()
```

```python
studf.notnull()
```

```python
studf
```

```python
# 筛选出没有空分数的所有行
studf.loc[studf["分数"].notnull(), :]
```

```python
# 删除全是空值的列
# how 有两个值 all: 这一列所有值为null删除, any: 这一列有一个null则删除
studf.dropna(axis=1, how="all", inplace=True)
```

```python
# 删除全是空值的行
studf.dropna(axis=0, how="all", inplace=True)
```

```python
studf
```

```python
# 将分数列为null的值填充为0分
studf.fillna({ "分数": 0 }) # 本地修改
# studf.loc[:, "分数"] = studf["分数"].fillna(0)
```

```python
# 填充姓名的缺失值
studf.loc[:,"姓名"] = studf["姓名"].fillna(method="ffill")
# studf
```

```python
studf
```

```python
# 将数据保存
studf.to_excel('./data/student_clean.xlsx', index=False)
```

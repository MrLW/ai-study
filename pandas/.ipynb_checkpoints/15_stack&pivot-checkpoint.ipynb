### 重塑和透视
#### what（是什么）
将列数据变成二维交叉形式，便于分析，这个过程叫做重塑或者透视.

#### how（如何用）

重塑或者透视目前已学的方式有两种：unstack 和 pivot，都是df的方法
* unstack: 对一个分层索引降级，将最外层index变成column
* pivot：接受三个参数，分别是index、column、和values，分别代表将横纵坐标以及值
    * 原理是对原始DF调用set_index创建分层索引，然后再调用unstack

#### where（在哪用）
在对数据统计分析时，我们有时候需要进行画图显示以便更加直观，但我们在读取excel时的数据基本无法直接画图（或者画出来的图并不直观），此时我们可以想到**重塑和透视** 对数据进行整理加工，然后再进行画图


```python
import pandas as pd 
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('./data/ml-25m/ratings.csv')
```

```python
# df.shape: (25000095, 4)
# 将大的csv 裁剪成 小的csv
df=df.head(100000)
```

```python
df.shape
```

```python
df.head(5)
```

```python
df['date'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
df.head(5)
```

```python
df['date'].dt.month
```

```python
# 实现数据统计(agg的用法)
# pv: 这是新的列
group=df.groupby([df['date'].dt.month, 'rating'])['userId'].agg(pv=np.sum)
```

```python
type(group)
```

```python
group.head(200)
```

对于上面的数据, 是没法根据月份、不同评分的次数来查看趋势

```python
group_stack=group.unstack()
```

```python
group_stack.head(5)
```

```python
group.plot()
```

```python
group_stack.plot()
```

#### pivot
其实是相当于对一个df使用set_index创建分层索引，然后调用unstack

```python
group.head()
```

```python
reset_group = group.reset_index()
```

```python
reset_group.head(3)
```

```python
pivot_group = reset_group.pivot(index="date",columns="rating", values="pv")
```

```python
pivot_group.plot()
```

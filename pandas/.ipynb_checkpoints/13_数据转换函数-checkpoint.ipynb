## 数据抓换函数

- map：只用于Series
- apply：用于Series每个值的处理，用于DataFrame每个轴的处理
- applymap：只用于DataFrame，实现对其每个值的处理

```python
import pandas as pd
df=pd.read_excel('./data/stocks/stock.xlsx')
```

```python
df["公司"].unique()
```

#### map

```python
dic={
    "阿里":"阿里巴巴",
    "拼多多":"来一刀",
    "百度":"纵里寻他千百度"
}
```

```python
df["公司1"]=df["公司"].map(dic)
```

```python
df["公司2"]=df["公司"].map(lambda x: dic[x])
```

```python
df.head()
```

### apply

```python
df["公司3"] = df["公司"].apply(lambda x: dic[x])
```

```python
df.head(4)
```

```python
df["公司4"]=df.apply(lambda col: dic[col["公司"]],axis=1)
```

```python
df.head(5)
```

#### applymap

```python
df.head()
```

```python
df[["收盘","开盘","高","低"]].applymap(lambda x: int(x))
```

```python
# 修改多列
df.loc[:,["收盘","开盘","高","低"]] =   df[["收盘","开盘","高","低"]].applymap(lambda x: int(x))
```
